```python
# Dependencies

from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
#url to scrape
url = 'https://www.nytimes.com/search?query=new+york+recycle'

#Retrieve page
response = requests.get(url)
#BeautifulSoup 
soup = BeautifulSoup(response.text, 'lxml')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('li', class_="SearchResults-item--3k02W")
#print(results)
for result in results:
    title = result.find('a').text
    
    print(title)
    print('-------------------')
```
